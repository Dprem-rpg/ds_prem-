```python
df1 = pd.read_csv('cleaned_feardata.csv')
df2 = pd.read_csv('historical_data.csv')
```

```python


# Convert 'Timestamp IST' in historical_df to datetime (format: "dd-mm-yyyy HH:MM")
df2['date'] = pd.to_datetime(df2['Timestamp IST'], format="%d-%m-%Y %H:%M")

```

```python
# Get min and max dates from the sentiment dataset
min_date = df1['date'].min()
max_date = df1['date'].max()

min_date

```

```python

max_date
```

```python


df1['date'] = pd.to_datetime(df1['date']).dt.normalize()
df2['Timestamp IST'] = pd.to_datetime(df2['Timestamp IST'], format="%d-%m-%Y %H:%M")
df2['date'] = df2['Timestamp IST'].dt.normalize()


df1.rename(columns={'value': 'sentiment_score'}, inplace=True)


df = pd.merge( df2,df1[['date', 'classification', 'sentiment_score']], on='date',how='left')



```

## EDA by Graph

```python
sns.countplot(x='classification', data=df3)
plt.title('Frequency of Market Sentiments')
plt.xlabel('Classification (1 = Fear, 5 = Extreme Greed)')
plt.ylabel('Number of Trades')
plt.show()
```

```python
df3.head()
```

```python
sns.countplot(x='Side',data=df3)
plt.title('Count of BUY/SELL')
```

```python

plt.figure(figsize=(10, 5))
sns.countplot(x='Direction', data=df3)
plt.xticks(rotation=45)  # Rotate x-axis labels
plt.title('Trade Direction Distribution')
plt.xlabel('Trade Direction')
plt.ylabel('Count')
plt.tight_layout() 
plt.show()
```

```python
sns.histplot(df3['Size USD'],bins=30)
plt.title("Distribution of USD")
```

```python
df3.columns
```

```python
#Bvariat analysis for using fast analysis we take 10000 data
df4 = df3.sample(n=10000, random_state=42)
```

```python
sns.countplot(x='Direction', hue='classification', data=df4)
plt.xticks(rotation=45)
plt.title('Trade Classification')
plt.xlabel('Trade Direction')
plt.ylabel('Count')
plt.tight_layout() 
plt.show()

```

```python
sns.scatterplot(x='Size USD', y='Closed PnL', hue='classification', data=df4)
plt.title('Size USD vs Closed PnL by Market Sentiment')
```

```python
sns.lineplot(x='classification', y='Execution Price', data=df3, marker='o')
plt.title("Average Execution Price vs Market Sentiment")
plt.xlabel("Market Sentiment (1 = Fear → 5 = Extreme Greed)")
plt.ylabel("Average Execution Price")
plt.grid(True)
plt.show()
```

## Feature selction

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df4.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Columns")
plt.show()
```
